```python
import tkinter as Tk
from tkinter import *

#function created

def btnclick(number):
    global val
    val=val+str(number)
    data.set(val)

def btnclear():
    global val
    val=""
    data.set(val)
    
def btnEquals():
    global val
    result=str(eval(val))
    data.set(result)

    
#CALL THE BUTTON IN FUNCTION TO WORK


root=Tk()
root.title("My Calculator")
root.geometry("400x500+500+150")
root.resizable(0,0)
val=""
data=StringVar()
display=Entry(root,textvariable=data,bd=40,bg="pink",font=("Ariel",20,'bold'),justify="right") 
display.grid(row=0,columnspan=4)
#btn7=Button(root,text="9",font=("Ariel",25,"bold"),bd=13,height=2,width=6)
#btn7.grid(row=1,column=0)
#FRIST ROW(7 ,8 ,9, * )
btn7=Button(root,text="7",font=("Ariel",15,"bold"),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(7))
btn7.grid(row=1,column=0)
btn8=Button(root,text="8",font=("Ariel",15,"bold"),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(8))
btn8.grid(row=1,column=1)
btn9=Button(root,text="9",font=("Ariel",15,'bold'),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(9))
btn9.grid(row=1,column=2)
btn_mul=Button(root,text="*",font=('Ariel',15,'bold'),bg="magenta",bd=12,height=2,width=6,command=lambda:btnclick('*'))
btn_mul.grid(row=1,column=3)
#SECOND ROW(4 ,5 ,6 , - )
btn4=Button(root,text="4",font=("Ariel",15,"bold"),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(4))
btn4.grid(row=2,column=0)
btn5=Button(root,text="5",font=('Ariel',15,"bold"),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(5))
btn5.grid(row=2,column=1)
btn6=Button(root,text="6",font=('Ariel',15,'bold'),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(6))
btn6.grid(row=2,column=2)
btn_sub=Button(root,text='-',font=('Ariel',15,'bold'),bg="magenta",bd=12,height=2,width=6,command=lambda:btnclick('-'))
btn_sub.grid(row=2,column=3)
#THRID ROW (1 ,2 , 3, + )
btn1=Button(root,text="1",font=("Ariel",15,"bold"),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(1))
btn1.grid(row=3,column=0)
btn2=Button(root,text="2",font=("Ariel",15,'bold'),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(2))
btn2.grid(row=3,column=1)
btn3=Button(root,text="3",font=("Ariel",15,"bold"),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(3))
btn3.grid(row=3,column=2)
btn_add=Button(root,text="+",font=('Ariel',15,'bold'),bg='magenta',bd=12,height=2,width=6,command=lambda:btnclick('+'))
btn_add.grid(row=3,column=3)
#FOURTH ROW (C ,0 , = , /)
btnc=Button(root,text="c",font=('Ariel',15,'bold'),bg="pink",bd=12,height=2,width=6,command=btnclear)
btnc.grid(row=4,column=0)
btn0=Button(root,text="0",font=("Ariel",15,'bold'),bg="cyan",bd=12,height=2,width=6,command=lambda:btnclick(0))
btn0.grid(row=4,column=1)
btn_equal=Button(root,text="=",font=("Ariel",15,'bold'),bg="pink",bd=12,height=2,width=6,command=btnEquals)
btn_equal.grid(row=4,column=2)
btn_divi=Button(root,text="/",font=("Ariel",15,'bold'),bg="magenta",bd=12,height=2,width=6,command=lambda:btnclick('/'))
btn_divi.grid(row=4,column=3)




root.mainloop()


```

```python
from tkinter import *
root=Tk() #object
root.title("Body Mass Index")
root.geometry("550x380+0+0")
root.resizable(1,1)
frame1=Frame(root,pady=10,bg='black',bd=16)
frame1.grid()
weight=StringVar()
height=StringVar()
result=StringVar()
frame2=Frame(frame1,width=450,height=170,pady=11,bg='red',bd=16,relief="sunken")
frame2.grid(row=0,column=0)
frame3=Frame(frame1,width=450,height=160,pady=11,bg='blue',bd=16,relief="sunken")
frame3.grid(row=1,column=0)
lblweight=Label(frame2,text="Enter Your Weight",font=("Ariel",16,"bold"),bg='yellow',bd=12)
lblweight.grid(row=0,column=0,sticky=W)
lblweight=Entry(frame2,textvariable=weight,font=("Ariel",16,"bold"),bg='cyan',bd=12)
lblweight.grid(row=0,column=1,sticky=W)
lblkg=Label(frame2,text="kg",font=("Ariel",16,"bold"),bg='cyan',bd=12)
lblkg.grid(row=0,column=2,sticky=W)
lblheight=Label(frame2,text="Enter Your Height",font=("Ariel",16,"bold"),bg='yellow',bd=12)
lblheight.grid(row=1,column=0,sticky=W)
lblheight=Entry(frame2,textvariable=height,font=("Ariel",16,"bold"),bg='yellow',bd=12)
lblheight.grid(row=1,column=1,sticky=W)
lblcm=Label(frame2,text="cm",font=("Ariel",16,"bold"),bg='cyan',bd=12)
lblcm.grid(row=1,column=2,sticky=W)
lblresult=Label(frame2,text="Show Result (BMI)",font=("Ariel",16,"bold"),bg='yellow',bd=12)
lblresult.grid(row=2,column=0,sticky=W)
lblresult=Entry(frame2,textvariable=result,font=("Ariel",16,"bold"),bg='cyan',bd=12)
lblresult.grid(row=2,column=1,sticky=W)
def add():
    w=float(lblweight.get())
    h=float(lblheight.get())/100 #coverting cm to meater
    r=w / (h * h)
    b=round (r,2)
    result.set(b)
def rset():
    weight.set("")
    height.set("")
    result.set("")
def Exit():
    root.destroy()
btnTotal=Button(frame3,text="Total",font=("Ariel",16,"bold"),bg='magenta',bd=12,pady=12,width=8,command=add)
btnTotal.grid(row=0,column=0)
btnReset=Button(frame3,text="Reset",font=("Ariel",16,"bold"),bg='magenta',bd=12,pady=12,width=8,command=rset)
btnReset.grid(row=0,column=1)
btnExit=Button(frame3,text="Exit",font=("Ariel",16,"bold"),bg='magenta',bd=12,pady=12,width=8,command=Exit)
btnExit.grid(row=0,column=2)
root.mainloop()
```
